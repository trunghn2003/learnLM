```python
import numpy as np
# Dữ liệu mẫu
names = np.array(['Linda', 'Tom', 'Sophia', 'James', 'Maria', 'John', 'Emma', 'Michael', 'Olivia', 'David'])
heights = np.array([1.65, 1.80, 1.55, 1.75, 1.62, 1.82, 1.68, 1.76, 1.58, 1.70])  # chiều cao (m)
weights = np.array([60, 75, 50, 68, 55, 85, 63, 78, 52, 72])                      # trọng lượng (kg)

# Tính toán chỉ số BMI
bmi = weights / (heights ** 2)
# Phân loại dựa trên BMI
categories = []

for value in bmi:
    if value < 18.5:
        categories.append("Underweight")
    elif 18.5 <= value <= 25:
        categories.append("Normal weight")
    else:
        categories.append("Overweight")
# In kết quả
for name, bmi_value, category in zip(names, bmi, categories):
    print(f"{name} có chỉ số BMI là {bmi_value:.2f} và được phân loại là {category}.")
```

```python
import matplotlib.pyplot as plt

# Data
data = {
    'Name': ['P1', 'P2'],
    'Height': [1.65, 1.70],  # in meters
    'Weight': [68, 80]  # in kilograms
}

# BMI calculation function
def calculate_bmi(weight, height):
    return weight / (height ** 2)

# Classification function
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

# Calculate BMI and classify
bmi_categories = {'Underweight': 0, 'Normal': 0, 'Overweight': 0, 'Obese': 0}
for i in range(len(data['Name'])):
    bmi = calculate_bmi(data['Weight'][i], data['Height'][i])
    classification = classify_bmi(bmi)
    bmi_categories[classification] += 1

# Plotting the distribution
labels = list(bmi_categories.keys())
values = list(bmi_categories.values())

plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('BMI Classification')
plt.ylabel('Number of Individuals')
plt.title('Distribution of BMI Classifications')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
file_path = 'bmi_dataset.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# BMI calculation function
def calculate_bmi(weight, height):
    return weight / (height ** 2)

# Classification function
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

# Calculate BMI and classify
bmi_categories = {'Underweight': 0, 'Normal': 0, 'Overweight': 0, 'Obese': 0}
bmi_values = []  # Store BMI values for plotting

for i in range(len(data)):
    bmi = calculate_bmi(data['Weight'][i], data['Height'][i])
    bmi_values.append(bmi)
    classification = classify_bmi(bmi)
    bmi_categories[classification] += 1

# 1. Bar Chart
labels = list(bmi_categories.keys())
values = list(bmi_categories.values())

plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('BMI Classification')
plt.ylabel('Number of Individuals')
plt.title('Distribution of BMI Classifications')
plt.show()

# 2. Pie Chart
plt.pie(values, labels=labels, colors=['blue', 'green', 'orange', 'red'], autopct='%1.1f%%')
plt.title('BMI Classification Distribution')
plt.show()

# 3. Histogram
plt.hist(bmi_values, bins=20, color='purple', edgecolor='black')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Distribution of BMI Values')
plt.show()

# 4. Box Plot
plt.boxplot(bmi_values)
plt.ylabel('BMI')
plt.title('Box Plot of BMI Values')
plt.show()

# 5. Scatter Plot (Weight vs Height)
plt.scatter(data['Height'], data['Weight'], c=bmi_values, cmap='viridis', edgecolor='black')
plt.colorbar(label='BMI')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight with BMI Color Scale')
plt.show()
```
